# Previsão de empréstimo
## Utilização de modelo Machine Learning

```python
# Upload dos dados de treino e teste

import pandas as pd

dados_treino = pd.read_csv('train_ctrUa4K.csv')

dados_teste = pd.read_csv('test_lAUu6dG.csv')
```

```python
display(dados_treino.head(5))
display(dados_teste.head(5))
```

```python
# Tratamento de dados faltantes

dados_treino['Credit_History'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 (Não) para valores faltantes
dados_treino['Dependents'].fillna(0, axis = 0, inplace = True) # Definindo 0 dependentes para valores faltantes
dados_treino['Married'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 (Não) para valores faltantes
dados_treino['LoanAmount'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 para valores faltantes
dados_treino['Loan_Amount_Term'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 para valores faltantes

dados_teste['Credit_History'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 (Não) para valores faltantes
dados_teste['Dependents'].fillna(0, axis = 0, inplace = True) # Definindo 0 dependentes para valores faltantes
dados_teste['Married'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 (Não) para valores faltantes
dados_teste['LoanAmount'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 para valores faltantes
dados_teste['Loan_Amount_Term'].fillna(0, axis = 0, inplace = True) # Definindo valor 0 para valores faltantes

dados_treino.dropna(axis = 0, inplace = True)
```

```python
# Reorganizando índices

dados_treino.index = range(len(dados_treino))
dados_teste.index = range(len(dados_teste))
```

```python
# Transformação dos dados representados por duas opções em valores binários [0/1]

dados_treino['Gender'] = dados_treino['Gender'].apply(lambda x: 0 if x == 'Male' else 1) # Male = 0 / Female = 1
dados_treino['Married'] = dados_treino['Married'].apply(lambda x: 0 if x == 'No' else 1) # Yes = 1 / No = 0
dados_treino['Education'] = dados_treino['Education'].apply(lambda x: 0 if x == 'Not Graduated' else 1) # Grad = 1 / Not Grad = 0
dados_treino['Self_Employed'] = dados_treino['Self_Employed'].apply(lambda x: 0 if x == 'No' else 1) # Yes = 1 / No = 0
dados_treino['Property_Area'] = dados_treino['Property_Area'].apply(lambda x: 0 if x == 'Urban' else 1 if x == 'Rural' else 2) # Urban = 0 / Rural = 1 / Semiurban = 2
dados_treino['Loan_Status'] = dados_treino['Loan_Status'].apply(lambda x: 0 if x == 'N' else 1) # Yes = 1 / No = 0
dados_treino['Dependents'] = dados_treino['Dependents'].apply(lambda x: 4 if x == '3+' else int(x))

dados_teste['Gender'] = dados_treino['Gender'].apply(lambda x: 0 if x == 'Male' else 1) # Male = 0 / Female = 1
dados_teste['Married'] = dados_treino['Married'].apply(lambda x: 0 if x == 'No' else 1) # Yes = 1 / No = 0
dados_teste['Education'] = dados_treino['Education'].apply(lambda x: 0 if x == 'Not Graduated' else 1) # Grad = 1 / Not Grad = 0
dados_teste['Self_Employed'] = dados_treino['Self_Employed'].apply(lambda x: 0 if x == 'No' else 1) # Yes = 1 / No = 0
dados_teste['Property_Area'] = dados_treino['Property_Area'].apply(lambda x: 0 if x == 'Urban' else 1 if x == 'Rural' else 2) # Urban = 0 / Rural = 1 / Semiurban = 2
dados_teste['Dependents'] = dados_treino['Dependents'].apply(lambda x: 4 if x == '3+' else int(x))
```

```python
# Salvando os IDs de teste antes de apagá-los do DataFrame original

IDs_teste = dados_teste['Loan_ID']
```

```python
dados_teste.drop('Loan_ID', axis = 1, inplace = True)
```

```python
# Tabelas depois de tratadas

display(dados_treino.head())
display(dados_teste.head())
```

## Aplicando o Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

modelo = RandomForestClassifier()
```

```python
# Separação das variáveis

y_treino = dados_treino['Loan_Status']
X_treino = dados_treino.drop(['Loan_ID', 'Loan_Status'], axis = 1)
```

```python
# Fit no modelo

modelo.fit(X_treino, y_treino)
```

```python
# Previsão dos dados de teste

previsoes = modelo.predict(dados_teste)
```

```python
# Score

modelo.score(dados_teste, previsoes)
```

```python
previsoes = pd.Series(previsoes)
```

```python
final_data = pd.concat([IDs_teste, previsoes], axis = 1)

final_data.rename(columns = {0: 'Loan_Status'}, inplace = True)

final_data['Loan_Status'] = final_data['Loan_Status'].apply(lambda x: 'Y' if x == 1 else 'N')
```
